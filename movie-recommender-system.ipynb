```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.head(1)
```

```python
# Columns to Keep: genres, id, keywords, title, overview, cast, crew
movies = movies[['movie_id', 'title', 'genres', 'keywords', 'overview', 'cast', 'crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]' --> format we have
# ['Action', 'Adventure', 'SciFi', 'Adventure'] --> Format we want
```

```python
import ast
```

```python
def convert(obj):
    List = []
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies.genres.apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    List = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            List.append(i['name'])
            counter += 1
        else:
            break
    return List
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    List = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            List.append(i['name'])
            break
    return List
```

```python
movies["crew"] = movies.crew.apply(fetch_director)
```
